```python
def h(t):
    return -1 * t**2 + 7 * t + 4


t = np.arange(0, 7.6, 0.1)

altura = h(t)

plt.plot(t, h(t), color="blue", label="altura pelo tempo")
plt.grid()
```

```python
def f1(x):
    return 3*x**2 + 4*x - 15

def f2(x):
    return np.exp(x/1)

x = np.arange(1, 5, .1)

y1 = f1(x)
y2 = f2(x)

plt.plot(x, y1, label='f1(x)')
plt.plot(x, y2, label='f2(x)')
plt.legend()
```
