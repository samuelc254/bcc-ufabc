# Organização de dados em planilhas

Planilhas são constituídas por **tabelas**, sendo
úteis para organizar dados, fazer cálculos com esses dados e
apresentar gráficos.
Vantagens:
- Facilita a visualização global dos dados.
- Identifica a posição de cada item de dado (linhas)
- Identifica os atributos dos dados (colunas)

- Armazenam itens de dados

As **colunas** da tabela:

OBS:

Exemplo: by=["Nota","Ano"]

ascending = variavel Booleana ou lista de variáveis Booleanas

## Ordenação com critério de desempate

**Note que "A Origem"	e "O Senhor dos Anéis" têm a mesma Nota 88, mas foram lançados em anos diferentes.**

Suponha que queiramos **ordenar por Nota (ordem crescente) e** utilizar como **critério de desempate o Ano (ordem crescente)**.

````
df.sort_values(by=["Nota","Ano"])
# ordena por Nota (ordem crescente) e utiliza como critério de desempate o Ano (ordem crescente)
````

````
df.sort_values(by=["Nota","Ano"], ascending=[False, True])
# ordena por Nota (ordem decrescente) e utiliza como critério de desempate o Ano (ordem crescente)
````

# Modificar a planilha ou criar uma cópia?
Quando usamos o sort_values, a planilha em df não é modificada.

Confira!

````
df.sort_values(by="Ano")
````

````
df
````

Quando executamos
```
df.sort_values(by="Ano")
```
* Criamos uma cópia de df e trabalhamos na cópia
* Podemos guardar essa cópia em outra variável, por exemplo, df_ordenada:

````
df_ordenada = df.sort_values(by = "Ano")
df_ordenada
````

Mas e se quiséssemos modificar a própria planilha ao invés de criar uma nova?

Podemos colocar a cópia na própria df:

````
df = df.sort_values(by="Ano")
df
````

Uma outra maneira é a seguinte, usando inplace=True :

````
df.sort_values(by="Ano", inplace=True)
df
````

### Qual a diferença entre essas duas opções?

### Por que temos duas maneiras de fazer a mesma atividade/tarefa?

Essas duas opções atingem o mesmo efeito: a planilha em df agora se encontra ordenada.

Mas os passos realizados são diferentes:

```df = df.sort_values(by="Ano")```
* Cria uma cópia de df
* Ordena a cópia de df
* Transfere a cópia para df

```df.sort_values(by="Ano", inplace=True)```
* Ordena a própria df sem retornar uma cópia

A primeira opção precisa de mais memória do computador por causa da cópia. Isso não é problema para planilhas pequenas, mas poderia ser se a planilha fosse muito grande.

# Filtros: comando `query("condição")`

Filtros são ferramentas muito importantes para a visualização de dados, pois nos **permitem ver apenas as entradas que satisfazem alguma condição**.

Veja df novamente:

````
df
````

Suponha que você queira implementar um filtro para verificar os **filmes com duração menor do que 125 minutos**. Se você executar o seguinte código, qual resultado  será apresentado?

````
df.query("Duração < 125")
````

### Exercício
Faça um filtro para visualizar apenas os filmes a partir do ano 2000.

Observação: Se quiser usar $\geq$ use ``>=`` Similar para $\leq$


````
# Sua resposta
df.query("Duração < 125 & Nota > 70")
````

E se quisermos olhar **apenas filmes do ano 2000**?


````
df.query("Ano == 2000")
````

E **apenas o filme Star Wars**? Observação: Para fazer uma condição de igualdade, use ``==`` e não ``=``

````
df.query("Título == 'Star Wars'")
````

Note que no comando acima 'Star Wars' aparece entre  **aspas simples**  (para indicar que é um texto).

E se quisermos olhar os **filmes com menos de 125 minutos a partir do ano 2000**?

Para isso combinamos as duas condições usando um ``and``:

````
df.query("Duração <125 and Ano >= 2000")
````



O ``and`` representa um "e lógico", quando queremos que as duas condições valham.

Ou seja, a condição 1 e a condição 2 valham ao mesmo tempo.

E se quiséssemos ver os filmes que satisfazem **pelo menos uma das condições**?

Por exemplo, você quer ver um **filme com menos de 125 minutos, mas, se for mais longo, você topa desde que ele seja a partir dos anos 2000**.

Nesse caso, usamos um ``or``, que representa um "ou lógico"

````
df.query("Duração < 125 | Ano >= 2000")
````

O "ou-lógico" é um pouco diferente do "ou" que usamos em Português usualmente.

O "ou-lógico" é o que chamamos de "ou-inclusivo" que significa que ou a primeira condição vale, ou a segunda condição vale, ou as duas valem.

### Exercício

Observação: Para fazer uma condição de igualdade, use ``==`` e não ``=``

````
# Sua resposta
df.query("Ano == 2001").sort_values(by="Nota")
````

## Agrupar - comando `groupby`

A ferramenta de agrupamento permite agregar dados e analisá-los, de acordo com algum dos campos.

**Suponha que você queira agrupar os filmes, considerando a sua nota e calcular a média da duração dos filmes em cada grupo.**

Por exemplo, Star Wars e The Matrix são os filmes no grupo de nota 87 e têm duração 121 e 136, respectivamente. Então, a média é (121+136)/2 = 257/2 = 128.5.

````
df.groupby("Nota")['Título'].count()
````

Acima estamos:
* Agrupando pela Nota
* Acessando o campo "Duração"
* Calculando as médias de Duração

Agora,
**suponha que você queira agrupar os filmes, considerando a sua nota e saber quantos filmes temos em cada grupo.**

````
df.groupby("Nota")["Título"].count()
````

Estamos:

* Agrupando pela Nota
* Acessando o campo "Título"
* Contando os títulos em cada grupo

**E se você quiser fazer uma maratona de filmes e assistir todos os filmes de cada ano? Suponha que você queira saber quanto tempo vai durar sua maratona para cada ano?**

Nesse caso, você quer:
* Agrupar pelo Ano
* Acessando o campo "Duração"
* Somar as durações em cada grupo

````
df.groupby("Ano")["Duração"].sum()
````

# Observação2: Planilhas csv que tenham a separação decimal de números reais marcada por vírgula [,] e não o ponto.



Vamos ver uma base de dados do **World Happiness Report 2019**.
O relatório pode ser encontrado [neste link.](https://worldhappiness.report/ed/2019/#read)  Downloads/Charpter 2: Online Data

Aqui podemos visualizar uma pequena parte desta tabela:


Captura de Tela 2021-03-01 às 06.44.10.png

Esta tabela usa o delimitador de dados ponto-e-vírgula ``;`` e o separador de casas decimais vírgula ``,``




Neste caso, ao invés de somente

df = pd.read_csv("link", sep = ';')

deve ser adicionar **decimal = ","** como o último parâmetro,

**para indicar** **para o Python que o separador DECIMAL na tabela é com VÍRGULA e não ponto** (o Python espera ponto). Assim, o comando fica:

df = pd.read_csv("link", sep=";", **decimal = ","**)


Veja:

````
#df = pd.read_csv("https://drive.google.com/u/3/uc?id=1CLOwB5IOrCugBi6Y2J_cqMnB7YDC8CSt&export=download", sep=";")
# Veja que este arquivo tem o separador DECIMAL com VÍRGULA e não ponto (o Python espera ponto). Portanto, adicionamos decimal="," como ultimo parametro na leitura da planilha:
df = pd.read_csv("https://drive.google.com/u/3/uc?id=1CLOwB5IOrCugBi6Y2J_cqMnB7YDC8CSt&export=download", sep=";",decimal=",")
#como a planinha é grande, vamos imprimir só as primeiras 9 linhas
df.head(9)
````

Note que na tabela criada acima, temos como separador de casas decimais ponto (e não vírgula, como no csv original). Python espera uma tabela que tenha ponto como separador de casas decimais.
