```python
import numpy as np

df = pd.read_csv("https://drive.google.com/u/3/uc?id=17rH9fRiOMttCwRLlVNysDhWNbstPoNlI&export=download", sep=";", decimal=",")
df.query("sexo == 'masculino'", inplace=True)
plt.plot(df["altura"], df["peso"], ".")
a, b = np.polyfit(df["altura"], df["peso"], 1)
x = np.linspace(1, 2.5, 100)
plt.plot(x, a * x + b)
print(a, b)
print(1.8 * a + b)
```
